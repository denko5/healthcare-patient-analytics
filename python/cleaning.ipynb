### FIXING COLUMN NAMES

```python
# ── Clean column names ─────────────────────────────────────────
# Strip whitespace from column names
df.columns = df.columns.str.strip()
```

```python
# Replacing the spaces with underscores and lowercasing everything
df.columns = df.columns.str.replace(' ', '_').str.lower()

print("\nCleaned column names:")
print(df.columns.tolist())
```

### CHECKING AND FIXING THE DATA TYPES

```python
# ── Fixing the date columns ───────────────────────────────────────────
df['date_of_admission'] = pd.to_datetime(df['date_of_admission'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
```

```python
# ── Checking if the numeric columns are correct ───────────────────────
df['age'] = pd.to_numeric(df['age'], errors='coerce')
df['billing_amount'] = pd.to_numeric(df['billing_amount'], errors='coerce')
df['room_number'] = pd.to_numeric(df['room_number'], errors='coerce')
df['length_of_stay'] = pd.to_numeric(df['length_of_stay'], errors='coerce')
```

```python
print("\nData types after fixing:")
print(df.dtypes)
```

### HANDLING MISSING VALUES

```python
# Checking for missing values
print("\n--- MISSING VALUES PER COLUMN ---")
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)
```

```python
missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})
print(missing_df[missing_df['Missing Count'] > 0])
```

### HANDLING DUPLICATES

```python
# Checking for any duplicates
print(f"\nDuplicate rows before: {df.duplicated().sum()}")
```

```python
# checking for logical duplicates such as (same patient, same admission date)
logical_dupes = df.duplicated(
    subset=['name', 'date_of_admission', 'billing_amount'], 
    keep='first'
)
print(f"Logical duplicates found: {logical_dupes.sum()}")
df = df[~logical_dupes]
```

```python
print(f"Duplicate rows after: {df.duplicated().sum()}")
print(f"Rows remaining: {len(df)}")
```

### DETECTING AND HANDLING OUTLIERS

```python
# Outlier detection using IQR method
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    print(f"  {column}: {len(outliers)} outliers | Range: {lower:.1f} – {upper:.1f}")  
    return lower, upper
```

```python
print("\n--- OUTLIER DETECTION ---")
age_lower, age_upper         = detect_outliers(df, 'age')
billing_lower, billing_upper = detect_outliers(df, 'billing_amount')
stay_lower, stay_upper       = detect_outliers(df, 'length_of_stay')
```

### DATA VISUALIZATION (EDA)

```python
# Exploratory Data Analysis (EDA) Plots
output_folder = r'C:\Users\Admin\Desktop\Healthcare-Project\reports'
os.makedirs(output_folder, exist_ok=True)
```

```python
sns.set_theme(style="whitegrid")
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Healthcare Dataset — Exploratory Analysis', 
             fontsize=16, fontweight='bold')
```

```python
# 1. Age distribution
axes[0, 0].set_title('Age Distribution')
sns.histplot(df['age'], bins=20, color='steelblue', ax=axes[0, 0])
axes[0, 0].set_xlabel('Age')
```

```python
# 2. Admissions by Medical Condition
axes[0, 1].set_title('Admissions by Medical Condition')
condition_counts = df['medical_condition'].value_counts()
sns.barplot(x=condition_counts.values, y=condition_counts.index, 
            palette='Blues_r', ax=axes[0, 1])
axes[0, 1].set_xlabel('Count')
```

```python
# 3. Admission Type breakdown
axes[0, 2].set_title('Admission Type')
admission_counts = df['admission_type'].value_counts()
axes[0, 2].pie(admission_counts, labels=admission_counts.index, 
               autopct='%1.1f%%', colors=['#2196F3','#4CAF50','#FF5722'])
```

```python
# 4. Billing Amount distribution
axes[1, 0].set_title('Billing Amount Distribution')
sns.histplot(df['billing_amount'], bins=30, color='coral', ax=axes[1, 0])
axes[1, 0].set_xlabel('Billing Amount (USD)')
```

```python
# 5. Length of Stay by Admission Type
axes[1, 1].set_title('Length of Stay by Admission Type')
sns.boxplot(x='admission_type', y='length_of_stay', 
            data=df, palette='Set2', ax=axes[1, 1])
axes[1, 1].set_xlabel('Admission Type')
axes[1, 1].set_ylabel('Days')
```

```python
# 6. Admissions over time (by month-year)
axes[1, 2].set_title('Monthly Admissions Trend')
monthly = df.groupby(['admission_year', 'admission_month']).size().reset_index(name='count')
monthly['period'] = monthly['admission_year'].astype(str) + '-' + monthly['admission_month'].astype(str).str.zfill(2)
monthly_sorted = monthly.sort_values(['admission_year', 'admission_month'])
axes[1, 2].plot(monthly_sorted['period'], monthly_sorted['count'], 
                marker='o', color='purple', linewidth=2)
axes[1, 2].tick_params(axis='x', rotation=45)
axes[1, 2].set_xlabel('Month')
axes[1, 2].set_ylabel('Admissions')
```

```python
plt.tight_layout()
plt.savefig(os.path.join(output_folder, 'eda_overview.png'), dpi=150, bbox_inches='tight')
plt.show()
print("\nEDA plot saved to reports folder.")
```
